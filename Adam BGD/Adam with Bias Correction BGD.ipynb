### Add colunms of one's to independent data

```python
X = np.concatenate((np.ones((len(x), 1)), x), axis=1)
```

## LR Full Implementation

### Step1: 
#### Initialize model parameters with zero.  
#### Choose value of the Learning rate. 

```python
# Initialize parameters (theta_0 , theta_1)
theta = np.zeros(len(X[1]))

# Choose the Learning rate
alpha = 0.06
```

### Step2: Predict the output h(x) = theta_0 + theta_1 * x

```python
def hypothesis(X, y, theta):
    h = np.dot(X , theta)
    e = h - y
    return e
```

### Step3: Calculate Cost function 𝑱(theta_0,theta_1 ).

```python
def cost_fun(e, m):
    J = (1/(2*m)) * np.dot(e.T , e)
    return J
```

### Step4: Calculate the gradient.

```python
def gradient(X, e, m):
    grad = (1/m) * np.dot(X.T , e)
    return grad
```

### Step5: Update the model parameters (simultaneously).

```python
def update(theta, m_th, v_th, epsilon, alpha):
    theta_new = theta - alpha * (1/(np.sqrt(v_th) + epsilon)) * m_th
    return theta_new
```

## Adam with Bias Correction BGD

```python
def BGD_Adam_BiasCorrection(X, y, theta, alpha, beta_1, beta_2, epsilon, num_itr, stopCriteria):
    
    theta_history = []
    J_history = []
    m = y.size

    cost = 0
    m_th = 0
    v_th = 0
    
    for i in range(1, num_itr+1):

        h = hypothesis(X, y, theta)

        j = cost_fun(h, m)
        if abs(cost-j) < stopCriteria:
            break
        J_history.append(j)
        cost = j
        
        grad_th = gradient(X, h, m)

        m_th = (beta_1 * m_th) + (1 - beta_1) * (grad_th)
        m_hat_th = m_th / (1 - beta_1**i)
        
        v_th = (beta_2 * v_th) + (1 - beta_2) * (grad_th**2)
        v_hat_th = v_th / (1 - beta_2**i)

        theta = update(theta, m_hat_th, v_hat_th, epsilon, alpha)

        theta_history.append(theta)

    return theta_history, J_history
```

## Use GD model to predict y values 

```python
theta_history, J_history = BGD_Adam_BiasCorrection(X ,y, theta, alpha, beta_1=0.1, beta_2=0.9, epsilon=1e-8, num_itr=500, stopCriteria=1e-2)
optimal_theta = theta_history[-1]
y_pred = X @ optimal_theta
```
